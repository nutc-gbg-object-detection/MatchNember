```python
from difflib import SequenceMatcher, get_close_matches
import sys

def search_myphton(myphone):
#     phone = correct_lis;
    phone = []
    for index,row in df.iterrows():
        phone.append(str(row['GET_TELEPHONE'])) #get telephone
        phone.append(str(row['SEND_TELEPHONE'])) #send telephone
    return_phone = []
   
    seq = SequenceMatcher(myphone)#a,b 帶入參數
    match = get_close_matches(myphone,phone,n=3,cutoff=0.3)
    if(len(match)>=1):
        #print('+-----------------+')
        for m in match:
            #print('|    '+(m)+'   |')
            #print('+=================+')
            return_phone += [m]
        return(return_phone)
    else: 
        return(-1)
        print("未匹配成功")
         
#     if(myphone=='quit'):
#         sys.exit();
```

```python
count_dist = {'1':0,'2':0,'3':0,'4':0,'total':0}
for index,item in enumerate(test_phone):
    #============start count send_phone=====
    
    for idx,value in enumerate(test_phone[item]):
        send_phone = test_phone[item]['send_phone']
        for val in test_phone[item]['send_data']:
            if send_phone in str(val):
                count_dist[str(val.index(send_phone)+1)] += 1
            else:
                count_dist['4'] += 1
            count_dist['total'] += 1
    #============start count get phone ======
    
    for idx, value in enumerate(test_phone[item]):
        get_phone = test_phone[item]['get_phone']
        for val in test_phone[item]['get_data']:
            if get_phone in str(val):
                count_dist[str(val.index(get_phone) + 1)] += 1
            else:
                count_dist['4'] += 1
            count_dist['total'] += 1
print(count_dist)
print('第一選項',int((count_dist['1']/count_dist['total'])*100),'%')
print('第二選項',int((count_dist['2']/count_dist['total'])*100),'%')
print('第三選項',int((count_dist['3']/count_dist['total'])*100),'%')
print('第四選項',int((count_dist['4']/count_dist['total'])*100),'%')
```
