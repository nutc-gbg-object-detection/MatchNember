```python
import random
def sampling(string,num):
    if(str(string) == 'nan' or str(string) == '0' or str(string) =='XX' or str(string)=='0  ' or str(string)=='0 '):
        return ''
    else:
        #========================start make random phone====================
        if num == 1:
            p = str(random.randint(0,9 if (len(string)-1) >= 10 else len(string)-1))
            c = str(random.randint(0,9))
        else:
            p = str(random.randint(0,9 if (len(string)-1) >= 10 else len(string)-1))+str(random.randint(0,9 if (len(string)-1) >= 10 else len(string)-1))
            c = str(random.randint(0,9))+str(random.randint(0,7))
        new = []
        for s in string:
            new.append(s)
        for index,item in enumerate(p):
            new[int(item)] = c[index]
        return ''.join(new)
sampling('66019191、66019192、66019194',2)

def Popstring(string,num):
    if(str(string) == 'nan' or str(string) == '0' or str(string) =='XX' or str(string)=='0  ' or str(string)=='0 '):
        return ''
    else:
        new = []
        for s in string:
            new.append(s)
        #========================start pop number====================
        if num == 1:
            new.pop(random.randint(0,len(string)-1))
            
        else:
            for i in range(num):
                popvalue = random.randint(0,len(string)-1)
                new.pop(popvalue)
                string = ''.join(new)
            
        return ''.join(new)
    
Popstring('0911333123',3)
```

```python
import pandas as pd

df = pd.read_excel('data/OP_Delivery_Requests.xlsx')

# print whole sheet data
df.head()
```

# use defflib  to search

```python
from difflib import SequenceMatcher, get_close_matches
import sys

#     phone = correct_lis;
phone = []
for index,row in df.iterrows():
    phone.append(str(row['GET_TELEPHONE'])) #get telephone
    phone.append(str(row['SEND_TELEPHONE'])) #send telephone
        
def search_myphton(myphone):
    return_phone = []
   
    seq = SequenceMatcher(myphone)#a,b 帶入參數
    match = get_close_matches(myphone,phone,n=3,cutoff=0.3)
    if(len(match)>=1):
        #print('+-----------------+')
        for m in match:
            #print('|    '+(m)+'   |')
            #print('+=================+')
            return_phone += [m]
        return(return_phone)
    else: 
        return(-1)
        print("未匹配成功")
         
#     if(myphone=='quit'):
#         sys.exit();
```

```python
if __name__ =='__main__':
    test_phone = {}
    for index,row in df.iterrows():
        
        test_dic = {}
        #==========================random get_telephone=========================
        test_dic['get_data'] = search_myphton(Popstring(row['GET_TELEPHONE'],4))
        
        test_dic['get_phone'] = str(row['GET_TELEPHONE'])
        #========================random send_telephone==========================

        test_dic['send_data'] = search_myphton(Popstring(row['SEND_TELEPHONE'],4))

        test_dic['send_phone'] = str(row['SEND_TELEPHONE'])
        test_phone[row['REQUEST_ID']] = test_dic
        
```

```python
for index,item in enumerate(test_phone):
    print(test_phone[item],'/r/n')
```

# The last count section

```python
count_dist = {'1':0,'2':0,'3':0,'4':0,'total':0}
for index,item in enumerate(test_phone):
    #============start count send_phone=====
    
    for idx,value in enumerate(test_phone[item]):
        send_phone = test_phone[item]['send_phone']
        if test_phone[item]['send_data']!=-1:
            testPhone_Array =  test_phone[item]['send_data']
            if send_phone in str(testPhone_Array):
                count_dist[str(testPhone_Array.index(send_phone)+1)] += 1
            else:
                    count_dist['4'] += 1
        else:
            count_dist['4'] += 1
        count_dist['total'] += 1
    #============start count get phone ======
    
    for idx, value in enumerate(test_phone[item]):
        get_phone = test_phone[item]['get_phone']
        if test_phone[item]['get_data']!=-1:
            getPhone_array = test_phone[item]['get_data']
            if get_phone in str(getPhone_array):
                count_dist[str(getPhone_array.index(get_phone) + 1)] += 1
            else:
                    count_dist['4'] += 1  
        else:
            count_dist['4'] += 1
        count_dist['total'] += 1
print(count_dist)
print('第一選項',int((count_dist['1']/count_dist['total'])*100),'%')
print('第二選項',int((count_dist['2']/count_dist['total'])*100),'%')
print('第三選項',int((count_dist['3']/count_dist['total'])*100),'%')
print('第四選項',int((count_dist['4']/count_dist['total'])*100),'%')
```
